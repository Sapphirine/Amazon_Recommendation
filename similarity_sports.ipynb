```python
import json
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import namedtuple
import datetime

def read_json_file(filename,meta_filename,take=lambda x:True):
    meta=dict()
    with open(meta_filename, 'r') as input_file:
        for line in input_file:
            #this is eval because it is not a standard json
            meta_content=eval(line)
            meta[meta_content['asin']]=meta_content
        
    Row = namedtuple('Row', ['ups', 'totals', 'overall', 'review', 'summary', 'reviewer_id', 'reviewer_name',
                            'review_time', 'product_id', 'brand', 'categories', 'price', 'product_title',
                            'review_year', 'review_month'])
    rows = []
    
    line_count = 0
    max_line_count = None
    with open(filename, 'r') as input_file:
        for line in input_file:
            review = json.loads(line)
            totals = review['helpful'][1]
            ups = review['helpful'][0]
            overall = float(review['overall'])
            review_text = review['reviewText']
            summary = review['summary']
            reviewer_id = review['reviewerID']
            reviewer_name = review.get('reviewerName', 'Unknown')
            review_time = int(review['unixReviewTime'])
            product_id = review['asin']
            # for product level info
            brand = meta[product_id].get('brand', 'Unknown')
            categories = meta[product_id].get('categories', [['Unknown']])
            price = float(meta[product_id].get('price', -999))
            product_title = meta[product_id].get('title', 'Unknown')
            dt = datetime.datetime.fromtimestamp(review_time)
            review_year = dt.year
            review_month = dt.month
            row = Row(ups=ups, totals=totals, overall=overall, review=review_text, summary=summary,
                            reviewer_id=reviewer_id, reviewer_name=reviewer_name,review_time=review_time,
                            product_id=product_id, brand=brand, categories=categories, price=price,
                            product_title=product_title, review_year=review_year, review_month=review_month)
            if take(row):
                rows.append(row)
            line_count += 1
            if max_line_count is not None and line_count >= max_line_count:
                break
    df = pd.DataFrame(rows, columns=Row._fields)
    return df
#meta_filename='C:/Users/Lenovo/AmazonProject/Amazon_Review_Helpfulness_Prediction-master/meta.json'
#filename='C:/Users/Lenovo/AmazonProject/Amazon_Review_Helpfulness_Prediction-master/reviews.json'
meta_filename='meta_Sports_and_Outdoors.json'
filename='reviews_Sports_and_Outdoors.json'
#meta_filename&filename should be updated
#filename_without_extension = os.path.splitext(os.path.basename(filename))[0]
data = read_json_file(filename, meta_filename)
#some fields in data may not be useful, later could be deleted, don't forget to update similarity matrix column
```

```python
def title_similarity(pair1_asin, pair2_asin):
    #pair1_asin, pair2_asin are input product_ids,eg.'0078800242','0439340039'
    temp1=data[data.product_id==pair1_asin]
    title1=temp1.iat[0,12]
    title1=title1.split()
    #print title1
    temp2=data[data.product_id==pair2_asin]
    title2=temp2.iat[0,12]
    title2=title2.split()
    #print title2
    if title1 == ['Unknown'] or title2 == ['Unknown']:
        title_similarity=0
    else:
        common=list(set(title1).intersection(title2))
        #print common
        union=list(set(title1).union(title2))
        #print union
        title_similarity=float(len(common))/len(union)
    return title_similarity

#title_similarity('0766233987','0974696153')#test
```

```python
def category_similarity(pair1_asin, pair2_asin):
    #pair1_asin, pair2_asin are input product_ids,eg.'0078800242','0439340039'
    temp1=data[data.product_id==pair1_asin]
    cat1=temp1.iat[0,10][0]
    #print cat1
    temp2=data[data.product_id==pair2_asin]
    cat2=temp2.iat[0,10][0]
    #print cat2
    common=list(set(cat1).intersection(cat2))
    #print common
    union=list(set(cat1).union(cat2))
    #print union
    category_similarity=float(len(common))/len(union)
    
    return category_similarity

#category_similarity('0766233987','0974696153')#test
```

```python
def brand_similarity(pair1_asin, pair2_asin):
    #pair1_asin, pair2_asin are input product_ids,eg.'0078800242','0439340039'
    temp1=data[data.product_id==pair1_asin]
    brand1=temp1.iat[0,9]
    brand1=brand1.split()
    #print brand1
    temp2=data[data.product_id==pair2_asin]
    brand2=temp2.iat[0,9]
    brand2=brand2.split()
    #print brand2
    if brand1 == ['Unknown'] or brand2 == ['Unknown']:
        brand_similarity=0
    else:
        common=list(set(brand1).intersection(brand2))
        #print common
        union=list(set(brand1).union(brand2))
        #print union
        brand_similarity=float(len(common))/len(union)
    return brand_similarity

#brand_similarity('0766233987','0974696153')#test
```

```python
def review_similarity(pair1_asin, pair2_asin):
    #pair1_asin, pair2_asin are input product_ids,eg.'0078800242','0439340039'
    temp1=data[data.product_id==pair1_asin]
    review1=temp1.iat[0,3]
    review1=review1.split()
    #print review1
    temp2=data[data.product_id==pair2_asin]
    review2=temp2.iat[0,3]
    review2=review2.split()
    #print review2
    common=list(set(review1).intersection(review2))
    #print common
    union=list(set(review1).union(review2))
    #print union
    review_similarity=float(len(common))/len(union)
    return review_similarity

#review_similarity('0766233987','0974696153')#test
```

```python
def price_similarity(pair1_asin, pair2_asin):
    #pair1_asin, pair2_asin are input product_ids,eg.'0078800242','0439340039'
    temp1=data[data.product_id==pair1_asin]
    price1=temp1.iat[0,11]
    #print price1
    temp2=data[data.product_id==pair2_asin]
    price2=temp2.iat[0,11]
    #print price2
    if price1 == -999 or price2 == -999:
        price_similarity=0
    else:
        price_similarity=1-float(abs(price1-price2))/(price1+price2)
    return price_similarity

#price_similarity('0766233987','0974696153')#test
```
